### import des outils

```python
import pandas as pd #pour lire, manipuler, analyser, nettoyer les donnees
import numpy as np  #pour faire des calculs scientifiques
import matplotlib as plt #pour visualiser les donnees
import seaborn as sns    #pour visualiser les donnees
import scipy as scp      #prendre des decisions basées sur les lois de probabilités
import sklearn as skl    #pour creer des modeles de machine learning (regression-classification S1)
```

```python
df=pd.read_csv("diamonds.csv")
df.head(3)
```

```python
X=df[['carat', 'table', 'depth','x', 'y', 'z']]
X.head(3)
```

```python
Y=df['price']
Y.head(3)
```

### Premier Modele IA 

```python
from sklearn.linear_model import LinearRegression #API
```

```python
modele=LinearRegression()  #On choisit un modele de regression
modele
```

```python
modele.fit(X,Y)  
#on entraine le modele pour qu'il puisse trouver y_train quand l'entree est X_train
```

```python
#Est ce que le modele s est bien entrainer
modele.score(X,Y)
```

```python
#Le modele est fiable en entrainement à 85.73% (precision ou R2 de 85.73%)
```

```python
coefficients = pd.Series(
    modele.coef_,
    index=X.columns
).sort_values(key=abs, ascending=False)

coefficients
```
